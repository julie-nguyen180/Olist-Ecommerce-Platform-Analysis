## Geolocation Table

```python
geo = pd.read_csv('/content/drive/MyDrive/Asm 5 BI/olist_geolocation_dataset.csv')
geo.info()
```

# Table Joining

```python
# Perform left join with orders on 'order_id'
ord_joined = ord_value.merge(orders, on='order_id', how='left')

# Perform left join with reviews on 'order_id'
ord_joined = ord_joined.merge(reviews, on='order_id', how='left')

# Perform left join with products on 'product_id'
ord_joined = ord_joined.merge(products, on='product_id', how='left')

# Perform left join with sellers on 'seller_id'
ord_joined = ord_joined.merge(sellers, on='seller_id', how='left')

# Perform left join with geo on zip_code prefix
# We use the 'seller_zip_code_prefix' from sellers and 'geolocation_zip_code_prefix' from geo
# Make sure the column names match for joining
ord_joined = ord_joined.merge(
    geo,
    left_on='seller_zip_code_prefix',
    right_on='geolocation_zip_code_prefix',
    how='left'
)
```

```python
ord_joined.info()
```

```python
# Check the number of null values for each column
null_counts = ord_joined.isnull().sum().sort_values(ascending=False)
print(null_counts)
```

```python
# Define the columns you want to check for nulls and drop rows with nulls
cols_to_drop_na = [
    'order_status',
    'order_purchase_timestamp',
    'geolocation_lat',
    'geolocation_lng',
    'geolocation_city',
    'geolocation_state',
    'geolocation_zip_code_prefix'
]

# Drop rows with any nulls in the specified columns
ord_cleaned = ord_joined.dropna(subset=cols_to_drop_na)
```

```python
# ord_cleaned.to_csv('ord_cleaned.csv', index=False)
```

```python
ord_cleaned.info()
```

# Aggregate Data by Seller & Month

```python
# Step 1: Extract order month
ord_cleaned['order_month'] = ord_cleaned['order_purchase_timestamp'].dt.to_period('M').astype(str)

# Step 2: Group and aggregate including unique customers
summary = ord_cleaned.groupby(['seller_id', 'order_month']).agg(
    total_orders=('order_id', 'nunique'),
    total_order_value=('order_value', 'sum'),
    avg_review_score=('review_score', 'mean'),
    unique_products_sold=('product_id', 'nunique'),
    total_customers=('customer_id', 'nunique'),
    geolocation_lat=('geolocation_lat', 'first'),
    geolocation_lng=('geolocation_lng', 'first'),
    geolocation_city=('geolocation_city', 'first'),
    geolocation_state=('geolocation_state', 'first'),
).reset_index()
```

```python
# Count product categories per seller-month
category_counts = (
    ord_cleaned.groupby(['seller_id', 'order_month', 'product_category'])
    .size()
    .reset_index(name='category_count')
)

# Get the most sold category per seller-month
top_categories = (
    category_counts.sort_values(['seller_id', 'order_month', 'category_count'], ascending=[True, True, False])
    .drop_duplicates(['seller_id', 'order_month'])
    .rename(columns={'product_category': 'top_category'})
    .drop(columns='category_count')
)
```

```python
# Merge top category into summary
seller_monthly_summary = summary.merge(top_categories, on=['seller_id', 'order_month'], how='left')
```
